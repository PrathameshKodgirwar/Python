# 10. Create a function to validate the password: password should contains at least one small letter, one capital letter, one number, and one special character except comma, quotes

```python
import re

def validate_password(pass1):
    if len(pass1) < 8:
        return "Password must be at least 8 characters long."

    if not re.search(r"[a-z]", pass1):
        return "Password must contain at least one lowercase letter."
    
    if not re.search(r"[A-Z]", pass1):
        return "Password must contain at least one uppercase letter."
    
    if not re.search(r"\d", pass1):
        return "Password must contain at least one number."
    
    if not re.search(r"[!@#$%^&*()_+=\-/\[\]{};:.<>?|]", pass1):
        return "Password must contain at least one special character (except comma, quotes)."
    
    if re.search(r"['\",]", pass1):
        return "Password should not contain commas or quotes."

    return "Password is valid."

pass1 = input("Enter a password: ")
print(validate_password(pass1))
```

# 11.Generate a decorator having logic to accept natural numbers . which can be used in a function of factorial

```python

def naturalNumRequired(func):
    def wrapper(n):
        if not isinstance(n, int) or n <= 0:
            return "Error: Input must be a natural number (positive integer)."
        return func(n)
    return wrapper

@naturalNumRequired
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

num = input("Enter a natural number: ")
try:
    num = int(num)
    print("Factorial:", factorial(num))
except ValueError:
    print("Error: Please enter a valid integer.")
```

# 12. Generate the decorator for a Garden visit to check if current time is between 7am to 7pm. If true the only call the function else generate the error.

```python
from datetime import datetime

def garden_visit_check(func):
    def wrapper(*args, **kwargs):
        current_hour = datetime.now().hour 
        
        if 7 <= current_hour < 19: 
            return func(*args, **kwargs)
        else:
            return "Error: Garden visit allowed only between 7 AM and 7 PM."
    
    return wrapper

@garden_visit_check
def visit_garden():
    return "Welcome to the garden! Enjoy your visit."

print(visit_garden())
```

# 15. Create a decorator to check if all input values are alpha numeric and based on it check PAN detail AAAAA1111X , rec.no first 3 chr next 5 number

```python
import re

def validate_inputs(func):
    def wrapper(*args, **kwargs):
        for arg in args:
            if not isinstance(arg, str) or not arg.isalnum():
                return "Error: All inputs must be alphanumeric."
            
        panPattern = r"^[A-Z]{5}\d{4}[A-Z]$"

        for arg in args:
            if re.match(panPattern, arg):  
                return func(arg)
            else:
                return "Error: Invalid PAN format. Expected AAAAA1111X."

    return wrapper

@validate_inputs
def check_pan_details(pan):
    return f"PAN {pan} is valid."

panNumber = input("Enter PAN number: ")
print(check_pan_details(panNumber))
```

# 17. Ask user to enter any 3 input e.g name, marks , mobile_num

if name is with other special characters then it should not not ask for next makrs input

note:

o Create separate functions for validation of every variable

o Every single block should have proper exception

o Every function must have the docstring

o Once the validation is done dictionary has to be created and store the dictionary to one txt file

```python
import re

def valid_name(name):
    """
    Validates the name to ensure it contains only alphabets and spaces.
    Returns True if valid, otherwise False.
    """
    if not name.isalpha() and " " not in name:
        print("Error: Name should contain only alphabets and spaces.")
        return False
    return True

def valid_marks(marks):
    """
    Validates marks to ensure it is a float or integer between 0 and 100.
    Returns True if valid, otherwise False.
    """
    try:
        marks = float(marks)
        if 0 <= marks <= 100:
            return True
        else:
            print("Error: Marks should be between 0 and 100.")
            return False
    except ValueError:
        print("Error: Marks should be a number.")
        return False

def valid_mobile(mobile):
    """
    Validates mobile number to ensure it is exactly 10 digits.
    Returns True if valid, otherwise False.
    """
    if re.fullmatch(r"\d{10}", mobile):
        return True
    else:
        print("Error: Mobile number must be exactly 10 digits.")
        return False

def get_input():
    """
    Gets user input for name, marks, and mobile number.
    If name is invalid, stops the process.
    Valid inputs are stored in a dictionary and written to a text file.
    """
    name = input("Enter Name: ").strip()
    
    if not valid_name(name):
        print("Process stopped due to invalid name.")
        return

    marks = input("Enter Marks: ").strip()
    if not valid_marks(marks):
        return
    
    mobile = input("Enter Mobile Number: ").strip()
    if not valid_mobile(mobile):
        return
    
    user_data = {"Name": name, "Marks": marks, "Mobile": mobile}

    with open("user_data.txt", "a") as file:
        file.write(str(user_data) + "\n")
    
    print("Data saved successfully:", user_data)

get_input()
```

# 18. Segregate file in small , medium and large size in different folder

o filesize less than and equal to 10KB is small file

o filesize more than 10 KB is the medium file

o filesize more than 40 kb is the large file

o create the log file

```python
import os
import shutil
import logging
from datetime import datetime

log_folder = "Sorted_Files"
os.makedirs(log_folder, exist_ok=True)
log_file = os.path.join(log_folder, "file_log.log")
logging.basicConfig(
    filename=log_file,
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S"
)

source_folder = "files_to_sort"
small_folder = os.path.join(log_folder, "Small")
medium_folder = os.path.join(log_folder, "Medium")
large_folder = os.path.join(log_folder, "Large")

os.makedirs(small_folder, exist_ok=True)
os.makedirs(medium_folder, exist_ok=True)
os.makedirs(large_folder, exist_ok=True)

def categorize_and_move_files():
    """
    Categorizes files based on their size and moves them into respective folders.
    Logs the details using the logging module.
    """
    if not os.path.exists(source_folder):
        logging.error(f"Source folder '{source_folder}' does not exist.")
        return

    files_moved = 0

    for filename in os.listdir(source_folder):
        file_path = os.path.join(source_folder, filename)

        if os.path.isfile(file_path):
            file_size_kb = os.path.getsize(file_path) / 1024  

            if file_size_kb <= 10:
                destination = small_folder
                category = "Small"
            elif file_size_kb <= 40:
                destination = medium_folder
                category = "Medium"
            else:
                destination = large_folder
                category = "Large"

            shutil.move(file_path, os.path.join(destination, filename))
            logging.info(f"Moved: {filename} | Size: {file_size_kb:.2f} KB | Category: {category}")
            files_moved += 1

    if files_moved == 0:
        logging.warning("No files were found to move.")
    else:
        logging.info(f"Total files moved: {files_moved}")

if __name__ == "__main__":
    categorize_and_move_files()
    print("File segregation complete! Check 'file_log.log' for details.")
```

# 19. In the hotel user is entering the details of menu, quantity and printing the bill. number of menu may varies with customer's order

generate a file having grocery bill as per the below format

```python
import os
import logging
from datetime import datetime

log_file = "hotel_bill.log"
logging.basicConfig(
    filename=log_file,
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S",
)

Hotel_Name = "Shri Hotel"
Gst_Rate = 0.10


def get_input():
    """Takes user input for menu items, quantity, and price. Returns a list of items."""
    items = []
    try:
        n = int(input("Enter the number of items: "))
        for i in range(n):
            menu = input(f"Enter name of item {i+1}: ").strip()
            quantity = int(input(f"Enter quantity of {menu}: "))
            price = float(input(f"Enter price of {menu}: "))
            items.append((menu, quantity, price))
    except ValueError as e:
        logging.error(f"Invalid input: {e}")
        print("Please enter valid numbers for quantity and price.")
        return get_input()
    return items


def generate_bill(items):
    """Generates a formatted bill and saves it to a text file."""
    total = sum(qty * price for _, qty, price in items)
    gst = total * Gst_Rate
    grand_total = total + gst

    receipt_no = datetime.now().strftime("%Y%m%d%H%M%S")


    bill_content = f"""
    |  {Hotel_Name}  |
    | {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}   Recpt: {receipt_no} |
    -------------------------------------
    | SR |   Menu   | Qnt | Price  |
    -------------------------------------
    """
    for i, (menu, qty, price) in enumerate(items, start=1):
        bill_content += f"| {i}  | {menu.ljust(8)} | {str(qty).rjust(3)} | {str(price * qty).rjust(5)} |\n"

    bill_content += f"""
    -------------------------------------
    | Total:        {total:.2f} |
    | GST 10%:     {gst:.2f} |
    | Grand Total: {grand_total:.2f} |
    -------------------------------------
    """

    with open("hotel_bill.txt", "w") as f:
        f.write(bill_content)
    
    logging.info(f"Bill generated. Total: {total:.2f}, GST: {gst:.2f}, Grand Total: {grand_total:.2f}")

    print("Bill generated successfully! Check 'hotel_bill.txt'.")


items = get_input()
generate_bill(items)
```

# 20. Ask user to enter the valid file path and check which word is having more repetition in the file

```python
import os
import logging
from collections import Counter

logging.basicConfig(
    filename="word_count.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S",
)

def get_valid_filepath():
    """
    Prompts the user for a valid file path and ensures the file exists.
    Returns:
        str: A valid file path
    """
    while True:
        file_path = input("Enter the file path: ").strip()
        if os.path.isfile(file_path):
            return file_path
        else:
            logging.error(f"Invalid file path entered: {file_path}")
            print("Invalid file path. Please enter a valid file path.")


def find_most_repeated_word(filePath):
    """
    Reads the file, counts word occurrences, and finds the most repeated word.
    Args:
        file_path (str): The path to the text file
    """
    try:
        with open(filePath, "r", encoding="utf-8") as file:
            text = file.read().lower() 

        words = text.split()
        word_counts = Counter(words)

        if not word_counts:
            logging.warning(f"The file '{filePath}' is empty or has no valid words.")
            print("The file is empty or contains no readable words.")
            return

        most_common_word, max_count = word_counts.most_common(1)[0] 

        print(f"The most repeated word is '{most_common_word}' (appears {max_count} times).")
        logging.info(f"Most repeated word in '{filePath}': '{most_common_word}' ({max_count} times).")

    except Exception as e:
        logging.error(f"Error processing file '{filePath}': {e}")
        print("An error occurred while processing the file. Please try again.")


file_path = get_valid_filepath()
find_most_repeated_word(file_path)
```

# 21.Create a function which will accept the word and the file path and returns the dictionary as {word: occurances}

```python
import os
import logging

logging.basicConfig(
    filename="word_occurrence.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S",
)

def count_word_occurrences(word, file_path):
    """
    Counts occurrences of a given word in the specified file.
    
    Args:
        word (str): The word to search for.
        file_path (str): The path to the text file.

    Returns:
        dict: {word: occurrences}
    """
    if not os.path.isfile(file_path):
        logging.error(f"Invalid file path: {file_path}")
        return {"error": "Invalid file path"}

    try:
        with open(file_path, "r", encoding="utf-8") as file:
            text = file.read().lower()

        word_count = text.split().count(word.lower())

        result = {word: word_count}
        logging.info(f"Word '{word}' found {word_count} times in '{file_path}'")
        return result

    except Exception as e:
        logging.error(f"Error reading file '{file_path}': {e}")
        return {"error": "An error occurred while reading the file"}


word = input("Enter the word to search: ").strip()
file_path = input("Enter the file path: ").strip()
result = count_word_occurrences(word, file_path)
print(result)
```

# 24. Ask user to enter the folder path. Folder may contain more than one file and with different extensions. Read only text file and return the more repetitive word

```python
import os
import logging
from collections import Counter

# Configure logging
logging.basicConfig(
    filename="most_repetitive_word.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S",
)


def find_most_repetitive_word(folder_path):
    """
    Finds the most repetitive word in all text files within a folder.

    Args:
        folder_path (str): Path of the folder containing text files.

    Returns:
        dict: {most_repetitive_word: count} or error message.
    """
    if not os.path.isdir(folder_path):
        logging.error(f"Invalid folder path: {folder_path}")
        return {"error": "Invalid folder path"}

    word_counter = Counter()

    try:
        for file_name in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file_name)

            if os.path.isfile(file_path) and file_name.endswith(".txt"):
                with open(file_path, "r", encoding="utf-8") as file:
                    words = file.read().lower().split()
                    word_counter.update(words)

        if not word_counter:
            logging.warning("No text files found or files are empty.")
            return {"error": "No text files found or files are empty"}

        most_common_word, frequency = word_counter.most_common(1)[0]

        logging.info(
            f"Most repetitive word '{most_common_word}' appeared {frequency} times."
        )
        return {most_common_word: frequency}

    except Exception as e:
        logging.error(f"Error processing files: {e}")
        return {"error": "An error occurred while processing files"}



folder_path = input("Enter the folder path: ").strip()
result = find_most_repetitive_word(folder_path)
print(result)
```

# 25. Create a class for file reading

```python
import os
import logging

logging.basicConfig(
    filename="file_reader.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S",
)

class FileReader:
    """
    A class to handle file reading operations.
    """

    def __init__(self, file_path):
        """
        Initializes the FileReader object with a file path.

        Args:
            file_path (str): The path of the file to read.
        """
        self.file_path = file_path

    def read_file(self):
        """
        Reads the entire content of the file.

        Returns:
            str: File content or error message.
        """
        if not os.path.isfile(self.file_path):
            logging.error(f"File not found: {self.file_path}")
            return "Error: File not found."

        try:
            with open(self.file_path, "r", encoding="utf-8") as file:
                content = file.read()
                logging.info(f"Successfully read file: {self.file_path}")
                return content
        except Exception as e:
            logging.error(f"Error reading file '{self.file_path}': {e}")
            return "Error: Unable to read file."

    def read_fileby_line(self):
        """
        Reads the file line by line and returns a list of lines.

        Returns:
            list: List of lines in the file or an error message.
        """
        if not os.path.isfile(self.file_path):
            logging.error(f"File not found: {self.file_path}")
            return ["Error: File not found."]

        try:
            with open(self.file_path, "r", encoding="utf-8") as file:
                lines = file.readlines()
                logging.info(f"Successfully read file line by line: {self.file_path}")
                return lines
        except Exception as e:
            logging.error(f"Error reading file '{self.file_path}': {e}")
            return ["Error: Unable to read file."]

file_path = input("Enter the file path: ").strip()
reader = FileReader(file_path)

content = reader.read_file()
print("\nFull File Content:\n", content)

lines = reader.read_fileby_line()
print("\nFile Content Line by Line:")
for line in lines:
    print(line.strip()) 
```

# 29. Create a class pharma. One user A will enter the medicine name and price. Another user B will generate the bill based on entry done by the first user

```python
import logging

logging.basicConfig(
    filename="pharma_error_log.log",
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S",
)


class Pharma:
    """
    A class to manage medicine entries and generate bills.
    """

    def __init__(self):
        """
        Initializes the Pharma class with an empty medicine database.
        """
        self.medicine_db = {}

    def add_medicine(self, name, price):
        """
        Adds a medicine and its price to the database.

        Args:
            name (str): Name of the medicine.
            price (float): Price of the medicine.

        Raises:
            ValueError: If the price is not a valid number.
        """
        try:
            price = float(price)
            self.medicine_db[name.lower()] = price
            print(f"Medicine '{name}' added successfully!")
        except ValueError:
            logging.error(f"Invalid price for medicine '{name}': {price}")
            print("Error: Invalid price entered. Please enter a valid number.")

    def generate_bill(self, order):
        """
        Generates a bill based on user B's order.

        Args:
            order (dict): Dictionary with medicine names as keys and quantities as values.

        Returns:
            float: Total bill amount.
        """
        total = 0
        missing_medicines = []

        print("\nBill Summary:")
        print("-" * 30)
        for med_name, qty in order.items():
            med_name = med_name.lower()
            if med_name in self.medicine_db:
                cost = self.medicine_db[med_name] * qty
                total += cost
                print(f"{med_name.capitalize()} x {qty} = ₹{cost:.2f}")
            else:
                missing_medicines.append(med_name)

        print("-" * 30)
        print(f"Total Amount: ₹{total:.2f}")

        if missing_medicines:
            missing_str = ", ".join(missing_medicines)
            logging.error(f"Medicines not found in database: {missing_str}")
            print(f"\nError: Medicines not found - {missing_str}")

        return total



pharma_store = Pharma()

print("\nUser A: Enter Medicines")
while True:
    med_name = input("Enter medicine name (or 'done' to stop): ")
    if med_name.lower() == "done":
        break
    med_price = input(f"Enter price for {med_name}: ")
    pharma_store.add_medicine(med_name, med_price)

    print("\nUser B: Place Order")
    order_dict = {}
    while True:
        med_name = input("Enter medicine name to buy (or 'done' to stop): ")
        if med_name.lower() == "done":
            break
        try:
            quantity = int(input(f"Enter quantity for {med_name}: "))
            order_dict[med_name] = quantity
        except ValueError:
            print("Error: Quantity must be a valid integer.")

    pharma_store.generate_bill(order_dict)
```
